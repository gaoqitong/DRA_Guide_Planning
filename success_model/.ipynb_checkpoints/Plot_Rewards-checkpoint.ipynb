```python
from __future__ import division
import tensorflow as tf
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
np.set_printoptions(threshold=np.nan)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = [25, 15]
matplotlib.rcParams.update({'font.size': 32})
```

```python
def smooth(input_data, coeff):
    output_data = range(len(input_data))
    for i in range(len(input_data)):
        if i-coeff < 0:
            output_data[i] = np.average(input_data[0:i+coeff])
        elif i+coeff > len(input_data) - 1:
            output_data[i] = np.average(input_data[i-coeff:])
        else:
            output_data[i] = np.average(input_data[i-coeff:i+coeff])
    return output_data
```

```python
small_dqn_file = pd.read_csv("./stats/DQN_Small.txt", delimiter=" |", header=None, engine="python")

small_dqn_epi = small_dqn_file[5]
small_dqn_reward = smooth(small_dqn_file[2], 5)
small_dqn_std = small_dqn_file[2].rolling(5).std()

small_guide_file = pd.read_csv("./stats/Guide_Small_A_T_NotC.txt", delimiter=" |", header=None, engine="python")

small_guide_epi = small_guide_file[5]
small_guide_reward = smooth(small_guide_file[2], 5)
small_guide_std = small_guide_file[2].rolling(5).std()
```

```python
f, ax = plt.subplots(1,1)
ax.plot(small_dqn_epi, small_dqn_reward, label="DQN w/o Guide", linewidth=5)
ax.fill_between(small_dqn_epi, small_dqn_reward-small_dqn_std, small_dqn_reward+small_dqn_std, alpha=0.35)

ax.plot(small_guide_epi, small_guide_reward, label="DQN w/ Guide", linewidth=5)
ax.fill_between(small_guide_epi, small_guide_reward-small_guide_std, small_guide_reward+small_guide_std, alpha=0.35)

ax.set_xlabel("# of Traning Episode")
ax.set_ylabel("Reward")
ax.legend()

f.tight_layout(rect=[0, 0, 1, 1])
f.savefig("medium_env_plot.png", dpi=200)

```
